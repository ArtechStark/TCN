```python
import numpy 
import os
import matplotlib.pyplot as plt
%matplotlib inline
font1 = {'family' : 'Times New Roman' , 'weight' : 'normal','size': 23}
font2 = {'family' : 'Times New Roman' , 'weight' : 'normal','size': 18}
```

```python
log = []
with open('../poly_music/0604_01.log', 'r') as f:
    log = f.read().splitlines()
log
```

```python
loss = [float(x.split(' ')[-1]) for x in log if 'Epoch' in x]

plt.figure(figsize=(20, 10))
iteration = np.arange(len(loss))
plt.plot(loss, color='red', label='training loss')
plt.legend(prop=font2)
plt.xlabel('iteration', font1)
plt.ylabel('train_loss', font1)
plt.show()
```

```python
loss = [float(x.split(' ')[-1]) for x in log if 'Test loss:' in x]

plt.figure(figsize=(20, 10))
iteration = np.arange(len(loss))
plt.plot(loss, color='red', label='training loss')
plt.legend(prop=font2)
plt.xlabel('iteration', font1)
plt.ylabel('test_loss', font1)
plt.show()
```
